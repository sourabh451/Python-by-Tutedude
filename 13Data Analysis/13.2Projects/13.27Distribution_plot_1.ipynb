```python
# DISTRIBUTION PLOT
import seaborn as sns
import numpy as np

np.random.seed(0)
x = np.random.randn(100)
ax_1 = sns.distplot(x)
```

```python
ax_2 = sns.distplot(x , rug = True , hist = False)
```

```python
ax_2 = sns.distplot(x , rug = False , hist = True)
```

```python
from scipy.stats import norm
ax_3 = sns.distplot(x , fit = norm)
```

```python
ax_3 = sns.distplot(x , fit = norm , kde = False)
```
