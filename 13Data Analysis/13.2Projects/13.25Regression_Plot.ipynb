```python
# REGRESSION PLOT
# vertical-axis = dependent
# horizontal-axis = independent
import seaborn as sns
flights = sns.load_dataset("flights")
print(flights)
```

```python
import matplotlib.pyplot as plt
sns.regplot(x = "year", y = "passengers", data = flights)
#plt.ylim(0,)
plt.ylim(0,500)
```
